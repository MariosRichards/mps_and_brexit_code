```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import gc
import re

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
%%time

# from https://www.theyworkforyou.com/mps/
mps = pd.read_csv("mps.csv"
                  , encoding = "ISO-8859-1")

mps.loc [mps["Constituency"]=='Ynys Môn', "Constituency"] = 'Ynys Mon'
mps.Constituency = mps.Constituency.apply(str.lower)


# https://medium.com/@chrishanretty/revised-estimates-of-leave-vote-share-in-westminster-constituencies-c4612f06319d#.dm91d4dwz
euRefVoteEstimates = pd.read_csv("Copy of Revised estimates of leave vote in Westminster constituencies - demog_based_estimates.csv"
                                 , encoding = "ISO-8859-1")

# I hate non-ascii characters!
euRefVoteEstimates.loc[ euRefVoteEstimates.Constituency=='Ynys MÃ´n', "Constituency"]= 'Ynys Mon'
euRefVoteEstimates.Constituency = euRefVoteEstimates.Constituency.apply(str.lower)

# Northern Ireland not included because not major UK parties *and* don't need approx figures because we have perfect figures
# http://www.bbc.com/news/uk-northern-ireland-36616830

# from https://www.theyworkforyou.com/mps/
results = pd.read_csv("RESULTS_2015.csv"
                  , encoding = "ISO-8859-1")
results["Constituency Name"] = results["Constituency Name"].apply(lambda x: str.lower(x) if not isinstance(x, float) else x)
# http://www.electoralcommission.org.uk/our-work/our-research/electoral-data
# results.loc[ results["Constituency Name"]=="Ynys Mon", "Constituency Name"] = 'Ynys Mon'
```

```python
merged_df = euRefVoteEstimates.merge(mps,"left","Constituency")
```

```python
useful_cols = ["Figure to use","Party"]
```

```python
merged_df[useful_cols].Party.astype('category').cat.categories
```

```python
merged_df["Party"] = merged_df["Party"].astype('category').cat.set_categories( ['Conservative', 'Labour', 'Scottish National Party',
       'Labour/Co-operative', 'Liberal Democrat', 'Plaid Cymru', 'Independent',
       'Green', 'Speaker', 'UKIP'] )

red_df = merged_df[useful_cols].copy()
```

```python
ax_list= red_df.hist(layout=(10,1),bins=np.linspace(0,1,20),by="Party", figsize=(10,20), sharex=True)
```

```python
red_df = merged_df.copy()

for drop_party in ['Liberal Democrat', 'Plaid Cymru', 'Independent', 'UKIP', 'Speaker', 'Green']:
    drop_ind = red_df[red_df.Party==drop_party].index
    red_df = red_df.drop(drop_ind)
    
red_df.Party[red_df.Party=="Labour/Co-operative"] = "Labour"    
red_df.Party = red_df.Party.cat.remove_unused_categories()  
red_df.Party = red_df.Party.cat.rename_categories(['Conservative', 'Labour+co_op', 'Scottish National Party'])
```

```python
# ["Labour/Co-operative"] -> Labour
```

```python
ax_list= red_df.hist(layout=(10,1),bins=np.linspace(0,1,20),by="Party", figsize=(10,20), sharex=True)
```

```python
marginality = pd.DataFrame(columns = ["constituency", "marginality"])
for constituencies in euRefVoteEstimates.Constituency.values:
    temp = results[ results["Constituency Name"]==constituencies ]["Share (%)"].sort_values(ascending=False)
    merged_df.loc[ merged_df["Constituency"]==constituencies, "marginality"] = temp.values[0] - temp.values[1]
    
```

```python
merged_df.loc[ merged_df["Constituency"]==constituencies, "marginality"] = temp.values[0] - temp.values[1]
```

```python
marginality.loc[1] = ["constituency", 56]
```

```python
merged_df["marginality"].hist(bins=20)
```

```python
most_marginal = (merged_df["marginality"] > merged_df["marginality"].quantile(0)) &\
(merged_df["marginality"] < merged_df["marginality"].quantile(.33))

middling = (merged_df["marginality"] > merged_df["marginality"].quantile(.33)) &\
(merged_df["marginality"] < merged_df["marginality"].quantile(.66))

least_marginal = (merged_df["marginality"] > merged_df["marginality"].quantile(.66)) &\
(merged_df["marginality"] < merged_df["marginality"].quantile(1))

larger_parties = red_df.Party.apply(lambda x: x not in ['Liberal Democrat', 'Plaid Cymru',
                  'Independent', 'UKIP', 'Speaker', 'Green'] )
```

```python
very_red_df = merged_df[useful_cols][least_marginal&larger_parties].copy()
very_red_df.loc[very_red_df.Party=="Labour/Co-operative","Party"] = "Labour"    
very_red_df.Party.cat.remove_unused_categories(inplace=True)  
very_red_df.Party.cat.rename_categories(['Conservative', 'Labour+co_op', 'Scottish National Party'],inplace=True)
ax_list= very_red_df.hist(layout=(10,1),bins=np.linspace(0,1,20),by="Party", figsize=(10,20), sharex=True)
```
